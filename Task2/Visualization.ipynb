Script created to visualizate the results of the model.

```python
history_TRAINED_load = np.load('history_TRAINED_model.npy',allow_pickle='TRUE').item()
```

```python
history_SCRATCH_load = np.load('history_SCRATCH.npy',allow_pickle='TRUE').item()
```

```python
#PLOT TRAINED MODEL EVOLUTION

fig = plt.figure(figsize=(14, 5))
# setting values to rows and column variables
rows = 1
columns = 2

# PLOT 1: summarize history for accuracy
fig.add_subplot(rows, columns, 1)
plt.plot(history_TRAINED_load['accuracy'])
plt.plot(history_TRAINED_load['val_accuracy'])
plt.title('Trained model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# PLOT 2: summarize history for loss
fig.add_subplot(rows, columns, 2)
plt.plot(history_TRAINED_load['loss'])
plt.plot(history_TRAINED_load['val_loss'])
plt.title('Trained model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
#PLOT SCRATCH MODEL EVOLUTION

fig = plt.figure(figsize=(14, 5))
# setting values to rows and column variables
rows = 1
columns = 2

# PLOT 3: summarize history for accuracy
fig.add_subplot(rows, columns, 1)
plt.plot(history_SCRATCH.history['accuracy'])
plt.plot(history_SCRATCH.history['val_accuracy'])
plt.title('Scratch model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# PLOT 4: summarize history for loss
fig.add_subplot(rows, columns, 2)
plt.plot(history_SCRATCH_load['loss'])
plt.plot(history_SCRATCH_load['val_loss'])
plt.title('Scratch model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
